```python
Temp = []
def graph(y):
    global Temp
    data_set = data.iloc[y].values
    Temp = [] #x_axis of data

    for i in range(len(data_set)):
        Temp.append(float(data_set[i]))
    title = str(Temp[0])
    Temp.remove(Temp[0])
    plt.plot(Temp,v_los)
    plt.title("For longitude:"+ title)
    plt.xlabel("Tb")
    plt.ylabel("V_LOS")
    plt.show()
    plt.savefig()
    return title
             
    
    
```

```python
for i in range(0,1024,32):
    graph(i)
    plt.savefig(graph(i)+".png")
```
