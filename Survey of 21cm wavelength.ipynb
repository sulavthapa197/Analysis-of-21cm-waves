```python
Temp = []
def graph(y):# Main function defined such that it can separate the required data as well as plot it 
    global Temp
    data_set = data.iloc[y].values
    Temp = [] #x_axis of data

    for i in range(len(data_set)):
        Temp.append(float(data_set[i]))
    title = str(Temp[0])
    img_name = title +".png"
    Temp.remove(Temp[0])
    plt.plot(Temp,v_los)
    plt.title("For longitude:"+ title)
    plt.xlabel("Tb / Kelvin")
    plt.ylabel("V_LOS/ kms-1")
    plt.savefig(img_name)
    plt.show()
   
    return title
             
    
    
```
